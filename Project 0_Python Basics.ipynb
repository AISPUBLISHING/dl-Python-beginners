# Python
Data containers are different data structures which we use to store data in various forms. Here, we will learn about data containers in python. It is necessary that we understand their use because the data is stored in those containers and we should know how to use them properly and when to use them.

## Data Containers

Here, we will discuss all three data containers: Lists, Tuples and Dictionaries. 

### Lists
List, as the name suggests, is a list of values. It contains items enclosed within square brackets. A difference between a regular array (in C++) and a list (in python) is that the items in a list can be of different types. For example following is a list.

````python
L=[24,3.5,'C','Myarray',23]                               # Define a list, containing objects of multiple types
print('List with different item types: ',L)
````

The elements in the list can be accessed through their positions(also called indexes). Indexing of elements in a list starts with 0 and goes to (length - 1). The slice operator [:] is used to select sublist from a bigger list. In slice operator, the first index shows the element of starting index. The range is open ended from end side i.e. the last value in slice operator is excluded from range. Some built in list functions are:
* **.append(val)**: Appends the value "val" at the end of list
* **.insert(index,val)**: Adds a new values "val" at index
* **.remove(val)**: Removes the first ocurrence of the value from the list
* **.pop()**: Removes and returns the last element from the list
* **.sort()**: Sorts the list

````python
print('Complete list L : ',L)
print('Element at index 3: ',L[3])
print('All elements from index 1 onwards: ',L[1:])
print('All elements before index 3(excluding index 3): ',L[:3])
print('Elements from index 2 to index 4: ',L[2:5])

# To repeat an element multiple times, we use * operator a number after * defines the number of times to repeat the list
print('Reptition of [4] 3 times: ',[4]*3)

# Now, we will modify the list elements:

print('\nUpdate list L: change the value of 3.5 to 6.5:')

# Assigning second element of the list a value of 6.5
L[1]=6.5

print('Complete list: ',L)

L.append(45)
print('\nAdding new element (45) at the end of L: ',L)

L.insert(3,'myname')
print('\nInserted a new element at index 3:',L)

L.remove('C')
print('\nAfter removing character \'C\' from the list:',L )

L.pop()
print('\nAfter popping last element from list:',L)
````

#### Arithmetic Operators on Lists
* \+ will combine two lists
* \* will repeat a list

````python
list1=[1,5,10,15]                                        # Define first list
list2=[2,4,8,12]                                         # Define second list
print('List 1:',list1)
print('List 2:',list2)
print('\nCombining two lists:',list1+list2)              # Combine two lists
print('\nRepeat list1 3 times:',list1*3)                 # Repeat a list 3 times
````

### Tuples
A tuple consists of a list of values enclosed within parentheses("()"). Similar to list, a tuple can have values of different types. Unlike lists, we can neither add/remove values from tuples nor we can update the values. Consider the same values from above example of a list and store them as a tuple:

````python
tupleseq=(24,3.5,'C','Myarray',23)                             # Defining a tuple containing multiple type objects
print('Complete tuple : ',tupleseq)
print('Element at index 3: ',tupleseq)
print('All elements from index 1 onwards: ',tupleseq[1:])     
print('All elements before index 3(excluding index 3): ',tupleseq[:3])
print('Elements from index 2 to index 4: ',tupleseq[2:5])
````

#### Arithmetic Operators on Tuples
* \+ will combine two tuples
* \* will repeat a tuple

````python
tuple1=('a','b')                                 # Defining first tuple
tuple2=('c','d')                                 # Defining second tuple
print('Combined Tuple: ',tuple1+tuple2)          # Combining two tuples
print('3 Times repeated Tuple: ',tuple1*3)       # Repeating a tuple 3 times
````

### Dictionaries

Dictionaries are associative arrays or hash tables and consist of values in {key : value} form. In other words, a dictionary maps a single key to a single value. The key or value can be any data type. Dictionaries are declared using curly braces and key-value pairs are defined inside. Dictionary container does not support arithemtic operations as lists and tuples do.

````python
#the value before colon (:) is the key and value after colon(:) is the mapped value
mydict={1:'one',
      2:'two',
      3:'three',
      4:'four'}

#We access values from a dictionary using its specific key. For example, 
print('Value against key 4:',mydict[4])
print('Value against key 2:',mydict[2])

#we can change the value of a key
mydict[2]='new value two'
print('New value against key 2:',mydict[2])

#we can add new key-value pair to the dictionary
mydict[7]='Seven'
print('Value of newly added key 7:',mydict[7])

#we can get all the keys and values of the dictionary
print('All keys of dictionary:',mydict.keys())
print('All values of dictionary:',mydict.values())
````

## Numpy Arrays:

A numpy array is actually a grid of same type of values. The array is indexed by a tuple of nonnegative integers. The shape of an array is a tuple of integers giving the size of the array along each dimension. We can initialize a numpy array from python list or list of lists. Consider following functions provided by numpy to initialize numpy arrays of different types:

* For a matrix filled with zeros: **zeros(shape)** where shape is a tuple of integers
* For a matrix filled with ones:  **ones(shape)** where shape is a tuple of integers
* For a random matrix with values between 0 and 1: **random.rand()**
* For a random integers matrix: **random.randint(low,high,shape)**, where low represents the lowest value, high represents the highest possible value for the array and shape is a tuple of integers
* We can get the shape(dimensions) of a numpy array using "shape" property: **nparray.shape**. This function returns (height, width, channel) for a 3 dimensional array.


````python
import numpy as np                                           # Import numpy library
zeros=np.zeros(shape=(5,5))                                  # A numpy array of shape (5 x 5) and filled with all zeros
print('zeros array:\n',zeros)

ones=np.ones(shape=(5,5))                                    # A numpy array of shape (5 x 5) and filled with all ones
print('\nones array:\n',ones)

# Numpy random.rand will have values between 0 and 1
randarr=np.random.rand(5,5)                                  # A random numpy array of shape (5 x 5) with values between 0 and 1
print('\nrandom array:\n',randarr)

#Numpy random int 
randintarr=np.random.randint(-3,3,(5,5))                     # A random integer numpy array of shape (5 x 5) with values between -3 and 3
print('\nrandom int array: \n',randintarr)
````

### Operations on a single Array
Now, We will learn about functions performed on a single numpy array.

#### Sum function
Numpy provides sum function to calculate sum of an array (all elements sum or axis wise sum):

```numpy.sum(A,axis)```
* A: Numpy array
* Axis: Axis along which to calculate the sum. If None or not defined, it will sum all the elements of array. For example, if axis = 0, sum will be calculated for each column and if axis = 1, sum will be calculated for each row.

````python
A=np.random.randint(-5,5,(3,4))                         # Create a random numpy array of shape (3 x 4) with values between -5 and 5
print('A:\n',A)
print('\nSum of all elements: ',np.sum(A))              # Sum of all values of array A
print('Sum along axis 0: ',np.sum(A,axis=0))            # Sum of all values of array A along the axis 0
print('Sum along axis 1: ',np.sum(A,axis=1))            # Sum of all values of array A along the axis 1
````

#### Min function
Numpy provides amin function to calculate minimum value in an array (for all elements or across an axis):

```numpy.amin(A,axis)```
* A: Numpy array
* Axis: Axis along which to find the minimum. If None or not defined, it will find minimum of all elements in the array. 

Similarly, other main functions in numpy are:
* **numpy.amax(A,axis)**: To find maximum value in an array across an axis
* **numpy.mean(A,axis)**: To find mean value in an array across an axis
* **numpy.median(A,axis)**: To find median value in an array across an axis

In all of the above functions, if axis is not provided, then the function is performed on the complete matrix (all elements).

````python
A=np.random.randint(low=0,high=255,size=(5,5))
print('A:\n',A)
print('\nMinimum of array  :',np.amin(A))
print('Maximum of array  :',np.amax(A))
print('Mean of array     :',np.mean(A))
print('Median of array   :',np.median(A))


print('\nMinimum of array at axis=0  :',np.amin(A,axis=0))
print('Maximum of array at axis=0  :',np.amax(A,axis=0))
print('Mean of array at axis=0     :',np.mean(A,axis=0))
print('Median of array at axis=0   :',np.median(A,axis=0))

print('\nMinimum of array at axis=1  :',np.amin(A,axis=1))
print('Maximum of array at axis=1  :',np.amax(A,axis=1))
print('Mean of array at axis=1     :',np.mean(A,axis=1))
print('Median of array at axis=1   :',np.median(A,axis=1))
````

### tile
Numpy's tile function replicates an array the number of times and dimensions given by reps

```numpy.tile(A,reps)```
* A: Numpy array
* reps: Number of times A will be repeated along each dimension

Consider following example where we repeat an array of 1 x 5 to make multiple rows of same values (tiling in dimension 0)

````python

# Generate a random numpy array of size (1 x 5) with values between -5 and 5
A= np.random.randint(low=-5,high=5,size=(1,5))                                 


print('A:\n',A)
print('\nA shape: ',A.shape)
tiled=np.tile(A,(5,1))                                   
# Tile(replicate) A, 5 times in dimension 0 and 1 times in dimension 1
print('\nTiled 5 times row wise(dimension 0) and 1 times column wise(dimension 1):\n',tiled)

tiled=np.tile(A,(5,2))
print('\nTiled 5 times row wise(dimension 0) and 2 times column wise(dimension 1):\n',tiled)
````

### concatenate
Numpy's concatenate function joins a sequence of arrays in a given dimension

```numpy.concatenate(A,axis)```

* A: A tuple of numpy arrays to be concatenated (A1,A2,A3...)
* axis: The axis number to join the arrays in

Consider a simple example below, where three numpy arrays of same dimensions, are combined together along rows (axis = 0).

````python
A= np.random.randint(low=-5,high=5,size=(1,5))                      # A random array of 1 x 5 (row x column) with values between -5 and 5
B= np.random.randint(low=-5,high=5,size=(1,5))                      # A random array of 1 x 5 (row x column) with values between -5 and 5
C= np.random.randint(low=-5,high=5,size=(1,5))                      # A random array of 1 x 5 (row x column) with values between -5 and 5

print('A:',A)
print('B:',B)
print('C:',C)

concatenated=np.concatenate((A,B,C),axis=0)                         # Concatenate 3 arrays along the 0th axis
print('\nConcatenated:\n',concatenated)
````

### dstack
Numpy's dstack function stacks a sequence of arrays in depth dimension. We will be using this function to deal with 3 channel (RGB) images.

```numpy.dstack(A)```

* A: A tuple of numpy arrays to be stacked (A1,A2,A3...)

Consider an example below, where three numpy arrays of same dimensions, are stacked together:

````python
R=np.random.randint(0,255,size=(24,24))
G=np.random.randint(0,255,size=(24,24))
B=np.random.randint(0,255,size=(24,24))
print('A shape:',R.shape)
print('B shape:',G.shape)
print('C shape:',B.shape)
stacked=np.dstack((R,G,B))
print('Stacked shape:',stacked.shape)
````

### unique
Numpy's unique function returns unique values in an array. 

```numpy.unique(A,return_index)```

* A: A numpy array
* return_index: If true, returns the indices of unique elements' first occurrence in original array


````python
A = np.random.randint(low=-2,high=2,size=(15,))
print('A:\n',A)
print('\nUnique from array:',np.unique(A))
uniquevals,indices=np.unique(A,return_index=True)
print('\nUnique from array with return indices: ',uniquevals,' - Indices in original array:',indices)
````

```pandas.DataFrame( data, index, columns, dtype)```

### Creating Datafame using Python List:
Let's create some dataframes from different types of data. First, we will create dataframe using a simple python list:

````python
import pandas as pd                                                                # Import pandas library
_list = [['car','truck'],['20000','30000']]                                        # Defined a List of lists
print ('Python list\n', _list)
print('\nDataframe:')
df = pd.DataFrame(_list)                                                           # create a dataframe from the list of lists
print (df)
````

### Creating Datafame using Numpy:
Now, we will create a dataframe using numpy array.

````python

import pandas as pd                                        # Import pandas library
import numpy as np                                         # Import numpy library

array = np.array(((2,3),(2,2)))                            # Create a numpy array
print ('Numpy array\n', array)

print('\nDataframe:')
df = pd.DataFrame(array)                                   # Create a dataframe using numpy array
print (df)                                                 # Display the dataframe
````

As you can see the indexes and columns are numbered from 0 to 1. Since we did not pass any index and column labels, dataframe is by default indexed and labeled using **np.arrange(n)**

````python

dictionary = {'Name':['Hassan', 'Muhammad', 'Ahmed', 'Ali'],'Age':[28,34,29,42]}         # A dictionary with keys defining column headings and arrays define their corresponding columns
print ('Dictionary\n',dictionary)

df = pd.DataFrame(dictionary)                                                            # Convert dictionary to dataframe
print (df)
````

## Accessing DataFrame Elements

Now we are going to see how can we extract rows, columns or comibnation of both from dataframes. The general rule for extracting a subset of a DataFrame

```DataFrame.loc[startrow:endrow, startcolumn:endcolumn]```

Lets see few examples how can we put this syantax to our use



````python

_list = [['Car',10000,100],['Truck',12000,200],['SUV',13000,300]]    # Create a dataframe

df = pd.DataFrame(_list,columns=['Vehicle','Price','Engine'])        # Assign Column names to dataframe

print (df)
````

We will use the general rule stated above to extract a subset of rows and columsn from this dataset

````python
df.loc[0:1,'Vehicle':'Price']                                        # Extract 0 - 1 rows and Vehicle : Price columns
````

### Accessing all values in a column
We can also access all the rows for any column as demonstrated in the following cell. We use (:) to specify all values in that dimension (in our case, dimension 0) and the column name to specify the column to extract values from.

````python
df.loc[:,'Vehicle']                                                 # Extract all values of column 'Vehicle'
````

### Accessing all values in a row
Similarly, we can access all values in a column for a specific row.

````python
df.loc[1,:]                                                  # Get all column values from row at index = 1
````

### Accessing a Single Element from Array
We can access a particular entry in the array by specifying the row and column value of element position.

````python
df.loc[2,'Vehicle']                                          # Access element at row = 2 and column='Vehicle'
````

## Updating Elements

We can update an element in dataframe just like we update elements in a python list. To update an element, we need to locate it by specifying its position (row,column) pair and then assigning the value for that position

````python
print('Before Updating element: ')
print(df)

df.loc[2,'Vehicle'] = 'sedan'                                # update second row of column 'Vehicle' to 'sedan'
print ('\nAfter updateing the element: \n',df)
````

# Matplotlib
To visualize data distributions, losses, accuracies or to even display images, we will use matplotlib in this tutorial. It is very intuitive and provides an easy way to draw plots, graphs, histograms, barcharts or scatter graphs.

## Drawing 2D Points
We will start with the fundamental part: drawing points on 2D cartesian coordinates. Consider these 5 points and let's draw them:
(1,5), (5,3), (11,-2), (8,13), (3,10)

````python
points=[(1,5), (5,3), (11,-2), (8,13), (3,10)]                         # Random 5 2D datapoints

# convert these points to numpy array
points=np.array(points)
print('points shape: ',points.shape)
````

2. Initialize a figure using figure() function

````python
import matplotlib.pyplot as plt                                       # Import pyplot module of matplotlib library
plt.figure()                                                          # Initialize a figure
plt.scatter(x,y)                                                      # scatter function will take two inputs: x values and their corresponding y values
plt.show()
````

```scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, verts=None, edgecolors=None, *, plotnonfinite=False, data=None, **kwargs)```

## Line Plotting
In matplotlib, a line is a sequence of coordinates joined together. Consider an equation of line:
**y = -13x + 7**. 
Now for a sequence of x values, we will have their respective y values for this line. To get a sequence of equidistant x values, numpy provides linspace function:

```linspace(start,stop,num,endpoint=True)```
* start: starting value of range
* end: end value of range
* num: Number of equally distant samples in range (start,stop)
* endpoint: If True, stop value is included in the range and if False, stop value is excluded.


````python
import numpy as np
x = np.linspace(start=-10,stop=10,num=50)                        # First, we generate a sequence of x values. 50 equally distant values between -10 and 10
y = -13 * x + 7                                                  # Get corresponding y values
````

Now we have our x values and their corresponding y values. To draw line, we will use plot function from pyplot:

```plot(x,y,fmt,label,linewidth)```
* x: x values
* y: corresponding y values
* fmt: A format string. Here we define the color, marker and border type of the line.
* label: A string value used as a label/name of the line. We can use this label to represent line in plot legends
* linewidth: A float value for the width of line.

Pyplot provides two more functions to specifically draw horizontal and vertical lines.
For horizontal line:

```axhline(y,**kwargs)```

* y: A float value on y axis to draw a horizontal line
* kwargs: to set color, line style, line width and other properties

For vertical line:

```axvline(x,**kwargs)```

* x: A float value on x axis to draw a vertical line
* kwargs: to set color, line style, line width and other properties


````python
import matplotlib.pyplot as plt

plt.figure()                                         # Initialize a figure
plt.title('A simple Line Plot')                      # Set title of figure
plt.plot(x,y,'r-',label='Line1',linewidth=2)         # Plotting the line
plt.xlabel('x axis',fontsize=12)                     # This will label X - axis of the plot
plt.ylabel('y axis',fontsize=12)                     # This will label Y - axis of the plot
plt.axhline(y=0,c='black',linestyle='--')            # This will draw a black horizontal line at y = 0
plt.axvline(x=0,c='black',linestyle='--')            # This will draw a black vertical line at x = 0 
plt.show()                                           # To show the plot
````

We can draw multiple lines and points on a single plot. Consider following example, where two lines and two set of 2D points are drawn. To differentiate between lines and set of points, we can assign different colors to each of them or we can use labels to caption in legends.

**Lengends**:

To add legend to our plot, we can call legend function from pyplot.

```legend(loc, **kwargs)```

* loc: This argument passed to legend function tells where to place the legend. There are 11 different positions available by default. We can pass the location using number or a string e.g. 'upper left' means to place the legend at the top left of plot.

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(start=-20,stop=20,num=1000)

y1 = 19*x + 119                                    # Defining lines and parabolas using equations
y2 = -13*x + 7
y3 = 4*x**2 - 113

# We will define two set of points
rand_points_set1=np.array([[10,800],[8,1200],[14,1050],[1,-100],[10,200]])
rand_points_set2=np.array([[9,173],[15,90],[2,-150],[0,-11],[7,-143]])

plt.figure()                                         # Initialize a figure
plt.title('Lines and Points Plot')                   # Set title of figure
plt.plot(x,y1,'r-',label='Line1',linewidth=2)        # Plotting first line
plt.plot(x,y2,'b-',label='Line2',linewidth=2)        # Plotting second line
plt.plot(x,y3,'g-',label='Parabola',linewidth=2)     # Plotting a quadratic function
plt.scatter(rand_points_set1[:,0],rand_points_set1[:,1],c='brown',label='Point set 1')
plt.scatter(rand_points_set2[:,0],rand_points_set2[:,1],c='orange',label='Point set 2')
plt.xlabel('x axis',fontsize=12)                     # This will label X - axis of the plot
plt.ylabel('y axis',fontsize=12)                     # This will label Y - axis of the plot
plt.axhline(y=0,c='black',linestyle='--')            # This will draw a black horizontal line at y = 0
plt.axvline(x=0,c='black',linestyle='--')            # This will draw a black vertical line at x = 0 
plt.legend(loc='upper left')                         # Place legend at top left corner of the plot
plt.show()                                           # To show the plot
````

## Subplots
Sometimes, we want figures to be drawn side by side, for comparison or for being relevant. Matplotlib provides this functionality as subplots. We can arranged images in grid or in linear form, each image is considered a subplot. Following is a simple example with 2 plots drawn side by side. We initialize subplots by calling "subplots" function and passing the number of rows and columns of images we want. The initialization returns two handles: one for complete figure and one for each axes. For example, as we create a figure with 2 subplots, it will return one handle for the whole figure and two axes handles through which we can access each subplot.

Some functions which we discussed above, change as:
* To set title of a subplot, we call "set_title()" on axes. To set title of complete figure, we call "suptitle" on figure handle returned by subplots function.

* We can access a subplot using its axes handle

* Rather than simple xlabel() and ylabel() functions, we have set_xlabel() and set_ylabel() functions for subplots with same parameters

* To set figure size, we use figure handle and call function set_size_inches and pass width and height values.

````python
import matplotlib.pyplot as plt
import numpy as np

# Divide numbers between -20 and 20 in 1000 equally distant values
x = np.linspace(start=-20,stop=20,num=1000)

# Define two plots (y1 and y2)
y1 = x - 3
y2 = x**2 + 2

# Define subplots
fig,ax = plt.subplots(1,2)                           # subplots: 1 row and 2 column (2 images side by side)
fig.set_size_inches(10.25, 5.25)

fig.suptitle('Plots',fontsize=14)                    # Set title of complete figure

ax[0].set_title('Line Plot')                         # Set title of axes 0 or subplot 1
ax[0].plot(x,y1,'r-',linewidth=2)                    # Plotting first line

ax[1].set_title('Parabola Plot')                     # Set title of axes 1 or subplot 1
ax[1].plot(x,y2,'b-',linewidth=2)                    # Plotting parabola
                                                     
for i in range(2):
    ax[i].set_xlabel('x-axis',fontsize=12)           # This will label X - axis of both subplots
    ax[i].set_ylabel('y-axis',fontsize=12)           # This will label Y - axis of both subplots
    ax[i].axhline(y=0,c='black',linestyle='--')      # This will draw a black horizontal line at y = 0
    ax[i].axvline(x=0,c='black',linestyle='--')      # This will draw a black vertical line at x = 0 
plt.show()                                           # To show the figure
````

Look at another example with 4 subplots:
As we now have 4 subplots shaped as a grid of 2 x 2, we need to access them using 2D indices.

````python
import matplotlib.pyplot as plt
import numpy as np

# Divide numbers between -20 and 20 in 1000 equally distant values
x = np.linspace(start=-20,stop=20,num=1000)

# Define 4 plots (y1,y2,y3,y4)
y1 = x - 3
y2 = x**2 + 2
y3 = -2*x
y4 = x**3 + 9

y=[y1,y2,y3,y4]

fig,ax = plt.subplots(2,2)                                 # subplots: 2 rows and 2 column (grid)
fig.set_size_inches(10.25, 10.25)                          # Set height and width of figure in inches
plt.subplots_adjust(top=0.9,wspace=0.4,bottom=0.15)        # To adjust spacing and gap between subplots
fig.suptitle('Plots',fontsize=18)                          # Set title of complete figure

handle_idx=0                                               # iterator to get y values
for i in range(2):
    for j in range(2):
        ax[i,j].plot(x,y[handle_idx],'r-',linewidth=2)     # Plot on axes
        ax[i,j].set_xlabel('x-axis',fontsize=12)           # This will label X - axis of subplot
        ax[i,j].set_ylabel('y-axis',fontsize=12)           # This will label Y - axis of subplot
        ax[i,j].axhline(y=0,c='black',linestyle='--')      # This will draw a black horizontal line at y = 0
        ax[i,j].axvline(x=0,c='black',linestyle='--')      # This will draw a black vertical line at x = 0 
        handle_idx+=1
plt.show()                                           # To show the figure
````

# OpenCV
OpenCV is an image processing library for python. We will be using this to load images, process images or to manipulate images for preprocessing. In this part, we will learn about basic functions from opencv as image loading, storing and image manipulation.

### Image Loading:
OpenCV's imread function reads an image file and returns a numpy array containing the pixel values. OpenCV reads the image in BGR format which means the order of the color channels is (Blue, Green, Red) as compared to alternative format RGB (Red, Green, Blue). In other words, we find Blue channel, Green channel and Red channel at indices 0, 1 and 2 respectively.

```imread(imagepath,loadflag)```

* imagepath: Complete path of image to load
* loadflag:  To specify the form to load the image in. The three options are: 
    * 1: To load a color image. Any transparency of image will be neglected. This is the default option.
    * 0: To load the image as a grayscale (one channel)
    * -1: To load the image with transparency (alpha channel)

The **imread** function loads the image as a numpy array.

Consider the example of reading an image:

````python
# First, import opencv library (named cv2)
import cv2                                              # importing opencv
im=cv2.imread('images/catimage.jpeg',1)                 # Reading image
print('Image shape:',im.shape)
````

The shape of image is (667,1000,3): height = 667, width = 1000, channels = 3. 

### Image Display
To display the loaded image, we will use a function provided by pyplot from matplotlib.

```imshow(image,cmap)```

* image: Numpy array or image loaded using opencv
* cmap: This parameter is used when the passed image is a grayscale (1 channel), otherwise it is ignored. It maps single scalar values in grayscale image to a color range. 

````python

import matplotlib.pyplot as plt                                # Import pyplot module from matplotlib library

plt.imshow(im)                                                 # Call imshow function of pyplot module of library to display the image
plt.show()                                                     # Call .show() function to actually display the image as a figure
````

As we can see, the image colors are not right. They are kind of inverted. The reason is because our loaded image is in BGR format i.e. our channels are sequences as (Blue, Green, Red) while matplotlib requires images in RGB format. There are two ways to transform our BGR image to RGB format:
1. Reorder the stack of channels in our image using numpy dstack
2. Use another function provided by opencv: **cvtColor**

We will use convert our image from BGR to RGB using both ways and will see our results. 

For the first point, we know that our image has dimensions (height, width, channel) and being a colored image, channels = 3. These channels are in BGR format. So, we will access each channel from image and will restack them to get image in RGB format. Then, we will use matplotlib to display our new image.

### Conversion from BGR to RGB using dstack
Since Red channel will be at the last index of our BGR Image, so to access it, we have to specify last index: im[:,:,2] will select red channel (in our case 3rd or last channel) of BGR image. Similarly, im[:,:,1] will select green channel (or 2nd channel) from BGR image and finally im[:,:,0] will select blue channel (or 1st channel) from BGR image. We will then stack them together in RGB format.

````python
import numpy as np
RGBformat=np.dstack((im[:,:,2],im[:,:,1],im[:,:,0]))            # Stacking channels to make RGB image
plt.imshow(RGBformat)                                           # Show image 
plt.show()
````

As you can see, our image is correctly displayed now. 

### Conversion from BGR to RGB using cvtColor
Numpy's cvtcolor function is used to change colorspace of an image. We can use it to convert our color format. 

```cvtColor(image,flag)```

* image: Numpy array image 
* flag:  specifying the type of conversion

There are several flags available for color space conversion but we are only looking for one that converts BGR to RGB: **cv2.COLOR_BGR2RGB**



````python
RGBimg=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)                           # OpenCV's cvtColor function converts image from BGR format to RGB format
plt.imshow(RGBimg)                                                  # Show RGB Image
plt.show()
````

### Conversion from BGR to Grayscale using cvtColor
In similar way, we can convert our RGB image to HSV color space or grayscale. For conversion to grayscale:

````python
grayscaleimg=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)     # Using OpenCV's cvtColor function, we convert BGR formatted image to Grayscale image.
plt.imshow(grayscaleimg,cmap='gray')                 # we assign gray cmap because by default, cmap has value = rc[image.cmap]
plt.show()
````

## Image Manipulation
Now, we will perform some operations on image to understand basic image processing which is usually done in data augmentation. For example, we will generate versions of an image with increased and decreased brightness as it can help model learn features in low light or in bright conditions. 

* To decrease brightness, we will divide the original pixel values by a number. This means, all the pixel values will decrease by same proportion and hence we will get darker image. Now consider a case when a pixel value is 230. Dividing it by 3 will give us a floating value. pyplot's imshow function can take either (0 - 1) float values or (0 - 255) integer values. Thus, it will not be a valid image for matplotlib. To cater this, we first need to normalize the image. Which means, we will first bring all the pixel values in (0 - 1) range and will then perform division. This way, we will remain in (0 - 1) range.


* To increase brightness, we will multiply the pixel values by some real number. It is possible that some values will increase beyond the maximum value for a pixel e.g. > 1.0 in our case. In this case, imshow function automatically clips those pixel values to highest possible (1.0).

````python
im=cv2.imread('images/lenna.png')                                  # Read image (images/lenna.png) using imread function of openCV
im=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)                              # Convert Image from BGR to RGB format

# Using subplots for image comparison
fig,ax=plt.subplots(1,3)
fig.set_size_inches(14.25, 8.25)

im=im/255.0                                                        # Normalize Image:

ax[0].imshow(im)                                                   # show image as a figure
ax[0].set_title('Original Image')

low_brightness=im/3                                                # This will divide all the pixel values by 3 (so the brightness will be reduced by a factor of 3)
ax[1].imshow(low_brightness)
ax[1].set_title('Lowered Brightness')

high_brightness = im*1.5                                           # The pixel values increase by a factor of 1.5 and hence, the brightness of image increases by same factor
ax[2].imshow(high_brightness)
ax[2].set_title('Increased Brightness')

plt.show()
````
